```python
# Keep it simple

from pathlib import Path
import os
import yaml
import pprint
import numpy as np
from influence_moo.config import load_config
from influence_moo.waves import sample_waves
from influence_moo.plotting import plot_grid, plot_pts, plot_vectors
```

```python
#alpha
cg = np.ones((7, 4))
cg[1:3, 1:3] = 0
cg[4:6, 1:3] = 0

# pois
pois = [
    {
        'position': [0.5, 3.5],
        'observation_radius': 1.0,
        'value': 1.0
    },
    {
        'position': [3.5, 3.5],
        'observation_radius': 1.0,
        'value': 1.0
    }
]

auvs = [
    {
        'path': [
            [3.0, 0.5],
            [0.5, 0.5],
            [0.5, 3.5]
        ],
    },
    {
        'path': [
            [6.5, 0.5],
            [3.5, 0.5],
            [3.5, 3.5]
        ]
    }
]

# asvs
asvs = [
    {
        'position': [3.5, 3.5]
    },
    {
        'position': [6.5, 3.5]
    }
]

for poi in pois:
    plot_pts(np.array([poi['position']]), None, '.', markersize=130, color='tab:green')

for auv in auvs:
    plot_pts(np.array(auv['path']), None, ':', color='purple')

for asv in asvs:
    plot_pts(np.array([asv['position']]), None, '^', color='orange', markersize=10)

plot_grid(cg, None, cmap='tab10_r')

print(3*'  '+'env:connectivity grid:')
cglist = cg.astype(int).tolist()
for row in cglist:
    print(4*'  '+'- '+str(row))
print(3*'  '+'env:auvs:')
for auv in auvs:
    print(4*'  '+'- path:')
    for position in auv['path']:
        print(5*'  '+f'- [{position[0]}, {position[1]}]')
print(3*'  '+'env:asvs:')
for asv in asvs:
    p = asv['position']
    print(4*'  '+f'- position: [{p[0]}, {p[1]}]')
print(3*'  '+'env:pois:')
for poi in pois:
    p = poi['position']
    o = poi['observation_radius']
    v = poi['value']
    print(4*'  '+f'- position: [{p[0]}, {p[1]}]')
    print(5*'  '+f'observation_radius: {o}')
    print(5*'  '+f'value: {v}')
```
