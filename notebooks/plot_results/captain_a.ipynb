```python
from pathlib import Path
import matplotlib.pyplot as plt
import os
import pandas as pd

root_dir = Path(os.path.expanduser("~/influence-multi-objective/results/captain_a/aug_26_2024/2_obs_rad/"))
dindirect_dir = root_dir/"D-Indirect"
g_dir = root_dir/"G"
```

```python
def get_nps(trials_dir):
    dind_trials = [str_ for str_ in os.listdir(trials_dir) if str_[:5]=="trial"]
    dind_trials = sorted(dind_trials, key=lambda x: int(x.split('_')[-1]))
    dind_dfs = [pd.read_csv(dindirect_dir/dind_trial/"fitness.csv") for dind_trial in dind_trials]
    dind_nps = [df["team_fitness_aggregated"].to_numpy() for df in dind_dfs]
    return dind_nps
```

```python
dind_nps = get_nps(dindirect_dir)
g_nps = get_nps(g_dir)
```

```python
for fit in dind_nps:
    plt.plot(fit)
plt.show()
```

```python
for fit in g_nps:
    plt.plot(fit)
plt.show()
```
